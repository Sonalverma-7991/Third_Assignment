```python
16.Create a code to check if a given list is sorted (either in ascending or descending order) or not.
Here’s a Python code to check if a given list is sorted in either ascending or descending order:

def is_sorted(lst):
    ascending = True
    descending = True
    
    # Check if the list is sorted in ascending order
    for i in range(len(lst) - 1):
        if lst[i] > lst[i + 1]:
            ascending = False
            break
    
    # Check if the list is sorted in descending order
    for i in range(len(lst) - 1):
        if lst[i] < lst[i + 1]:
            descending = False
            break
    
    # If either ascending or descending is True, the list is sorted
    if ascending:
        return "The list is sorted in ascending order."
    elif descending:
        return "The list is sorted in descending order."
    else:
        return "The list is not sorted."

# Example usage
lst1 = [1, 2, 3, 4, 5]
lst2 = [5, 4, 3, 2, 1]
lst3 = [3, 1, 4, 2]

print(is_sorted(lst1))  # The list is sorted in ascending order.
print(is_sorted(lst2))  # The list is sorted in descending order.
print(is_sorted(lst3))  # The list is not sorted.
Explanation:
The function is_sorted() checks if the list is sorted by iterating through the list twice.
The first loop checks if the list is in ascending order.
The second loop checks if the list is in descending order.
If the list is found to be sorted in either direction, the corresponding message is returned.
If neither condition is met, the function concludes that the list is not sorted.
```

```python
17.Write a code to merge two sorted lists into a single sorted list.

Here’s a Python code to merge two sorted lists into a single sorted list:

def merge_sorted_lists(list1, list2):
    # Initialize pointers for both lists
    i, j = 0, 0
    merged_list = []
    
    # Traverse both lists and merge them
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1
    
    # If there are remaining elements in list1, add them to merged_list
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1
    
    # If there are remaining elements in list2, add them to merged_list
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1
    
    return merged_list

# Example usage
list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]

result = merge_sorted_lists(list1, list2)
print("Merged sorted list:", result)
Explanation:
Two pointers i and j are initialized to traverse list1 and list2 respectively.
The elements from both lists are compared, and the smaller element is added to the merged_list.
The pointers are incremented accordingly to continue merging.
After exiting the main loop, any remaining elements in either list are appended to merged_list.
Running this example will give you:

Merged sorted list: [1, 2, 3, 4, 5, 6, 7, 8]
```

```python
20.Write a code to shuffle a given list randol
 without using 'n
 built-in shuffle functions.
Here’s a Python code to shuffle a given list randomly without using any built-in shuffle functions:

import random

def custom_shuffle(input_list):
    # Get the length of the list
    length = len(input_list)
    
    # Iterate over the list
    for i in range(length):
        # Generate a random index
        rand_index = random.randint(0, length - 1)
        
        # Swap the current element with the element at the random index
        input_list[i], input_list[rand_index] = input_list[rand_index], input_list[i]

# Example usage
my_list = [1, 2, 3, 4, 5]
print("Original list:", my_list)

custom_shuffle(my_list)
print("Shuffled list:", my_list)
Explanation:
The code manually shuffles the list by iterating over each element and swapping it with a randomly chosen element.
The random.randint(0, length - 1) function generates a random index within the valid range of indices for the list.
This method effectively shuffles the list in place, giving a random permutation each time it's run.
```

```python
31.Write a code that takes a list of words as input and returns a dictionary where the keys are unique words
and the values are the frequencies of those words in the input list.

Here's a Python code snippet that takes a list of words as input and returns a dictionary where the keys are unique words and the values are the frequencies of those words in the input list:

def word_frequencies(word_list):
    frequency_dict = {}
    
    for word in word_list:
        if word in frequency_dict:
            frequency_dict[word] += 1
        else:
            frequency_dict[word] = 1
    
    return frequency_dict

# Example usage:
words = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']

frequencies = word_frequencies(words)
print("Word Frequencies:", frequencies)
Output:
css
Word Frequencies: {'apple': 3, 'banana': 2, 'orange': 1}
Explanation:
The function word_frequencies takes a list of words (word_list) as input.
It initializes an empty dictionary called frequency_dict.
It then iterates through each word in the list:
If the word is already in the dictionary, it increments the count associated with that word.
If the word is not in the dictionary, it adds the word as a key with a value of 1.
The function returns the frequency_dict, which contains the unique words as keys and their frequencies as values.
```

```python
32.Write a code that takes two dictionaries as input and merges them into a single dictionary. If there are
common keys the values should be added together.
Here's a Python code snippet that merges two dictionaries. If the dictionaries have common keys, the values associated with those keys will be added together:

def merge_dictionaries(dict1, dict2):
    merged_dict = dict1.copy()  # Start with a copy of the first dictionary
    
    for key, value in dict2.items():
        if key in merged_dict:
            merged_dict[key] += value  # Add the values if the key is common
        else:
            merged_dict[key] = value  # Otherwise, just set the value
    
    return merged_dict

# Example usage:
dict1 = {
    'a': 5,
    'b': 3,
    'c': 7
}

dict2 = {
    'b': 2,
    'c': 10,
    'd': 8
}

merged_dict = merge_dictionaries(dict1, dict2)
print("Merged Dictionary:", merged_dict)
Output:
css
Merged Dictionary: {'a': 5, 'b': 5, 'c': 17, 'd': 8}
Explanation:
The function merge_dictionaries takes two dictionaries dict1 and dict2 as input.
It starts by creating a copy of dict1 to ensure the original dictionary remains unaltered.
It then iterates over the items in dict2:
If a key from dict2 exists in merged_dict, the function adds the value from dict2 to the existing value in merged_dict.
If the key does not exist in merged_dict, the key-value pair from dict2 is added to merged_dict.
Finally, the function returns the merged dictionary.
```

```python
33. Write a  code to access a  value  in a nested dictionary. The function should take the dictionary and a list of
keys as input and return the corresponding value. If any of the keys do not exist in the dictionary the
function should return None .
Here's a Python code snippet that accesses a value in a nested dictionary. The function takes the dictionary and a list of keys as input and returns the corresponding value. If any of the keys do not exist, the function returns None:

def get_nested_value(d, keys):
    for key in keys:
        if isinstance(d, dict) and key in d:
            d = d[key]
        else:
            return None
    return d

# Example usage:
nested_dict = {
    'a': {
        'b': {
            'c': 42
        }
    },
    'x': {
        'y': {
            'z': 99
        }
    }
}

# Accessing an existing value
value = get_nested_value(nested_dict, ['a', 'b', 'c'])
print("Value:", value)  # Output: 42

# Trying to access a non-existing value
value = get_nested_value(nested_dict, ['a', 'b', 'd'])
print("Value:", value)  # Output: None

# Trying to access a value in a non-existing path
value = get_nested_value(nested_dict, ['a', 'x', 'y'])
print("Value:", value)  # Output: None
Explanation:
The function get_nested_value takes two arguments:
d: The nested dictionary.
keys: A list of keys representing the path to the value.
The function iterates over the list of keys:
If the current level is a dictionary and the key exists, it moves deeper into the nested dictionary.
If any key does not exist or the current level is not a dictionary, the function returns None.
If all keys are found, the function returns the value at the end of the path.
```

```python
34.Write  a  code  that  takes a dictionary as input and returns a sorted version of it based on the values. You
can choose whither to sort in ascending or descending order.

Here's a Python code snippet that sorts a dictionary based on its values, with an option to sort in ascending or descending order:

def sort_dictionary_by_values(d, ascending=True):
    sorted_dict = dict(sorted(d.items(), key=lambda item: item[1], reverse=not ascending))
    return sorted_dict

# Example usage:
original_dict = {
    'a': 5,
    'b': 2,
    'c': 9,
    'd': 3,
    'e': 7
}

# Sort in ascending order
sorted_dict_ascending = sort_dictionary_by_values(original_dict, ascending=True)
print("Ascending order:", sorted_dict_ascending)

# Sort in descending order
sorted_dict_descending = sort_dictionary_by_values(original_dict, ascending=False)
print("Descending order:", sorted_dict_descending)
Output:
css
Ascending order: {'b': 2, 'd': 3, 'a': 5, 'e': 7, 'c': 9}
Descending order: {'c': 9, 'e': 7, 'a': 5, 'd': 3, 'b': 2}
Explanation:
The function sort_dictionary_by_values takes two arguments:
d: The original dictionary to be sorted.
ascending: A boolean flag indicating whether to sort in ascending (True) or descending (False) order.
The dictionary is sorted using the sorted() function, which sorts the items based on the values (item[1]).
The reverse parameter is set to not ascending, so if ascending is True, the dictionary will be sorted in ascending order, otherwise, in descending order.
The sorted items are then converted back into a dictionary using the dict() constructor.
```

```python
35.. Write a code that inverts a dictionary swapping keys and values. Ensure that the inverted dictionary
correctly  handles cases where multiple keys have the same value by storing the keys as a list in the
inverted dictionary.
Here's a Python code snippet that inverts a dictionary, handling cases where multiple keys have the same value by storing the keys as a list in the inverted dictionary:

def invert_dictionary(d):
    inverted_dict = {}
    
    for key, value in d.items():
        if value in inverted_dict:
            inverted_dict[value].append(key)
        else:
            inverted_dict[value] = [key]
    
    return inverted_dict

# Example usage:
original_dict = {
    'a': 1,
    'b': 2,
    'c': 1,
    'd': 3,
    'e': 2
}

inverted_dict = invert_dictionary(original_dict)
print(inverted_dict)
Output:
css
{1: ['a', 'c'], 2: ['b', 'e'], 3: ['d']}
Explanation:
The function invert_dictionary iterates through the original dictionary d.
For each key, value pair in the dictionary:
If the value is already a key in the inverted_dict, the function appends the key to the list associated with that value.
If the value is not a key in the inverted_dict, the function creates a new entry with the value as the key and initializes the value as a list containing the key.
```
